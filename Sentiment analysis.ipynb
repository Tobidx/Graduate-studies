```python
import pandas as pd
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import OneHotEncoder
from scipy.sparse import hstack
from textblob import TextBlob

file_path = '/Users/ajibolaoluwatobiloba/Desktop/AMOD5410- Big Data/Assignments/Assignment 3 --- ML & Spark/medical_records.csv'
data = pd.read_csv(file_path)

data.drop(['patient_id','date_of_birth'], axis=1, inplace=True)
data.dropna(inplace=True)

categorical_values = ['gender', 'medications', 'medical_conditions', 'allergies']

encoded_features = []
for val in categorical_values:
    onehot_encoder = OneHotEncoder(sparse_output=True)
    encoded_column = onehot_encoder.fit_transform(data[[val]])
    encoded_features.append(encoded_column)
encoded_features = hstack(encoded_features)

n_clusters = 4
kmeans = MiniBatchKMeans(n_clusters=n_clusters, random_state=50)
kmeans.fit(encoded_features)
cluster_labels = kmeans.labels_
data['cluster'] = cluster_labels
cluster_centers = kmeans.cluster_centers_
print("Cluster centers:")
print(cluster_centers)

text_columns = ['medical_conditions','medications','allergies']
for col in text_columns:
    data[f'{col}_sentiment'] = data[col].apply(lambda x: TextBlob(str(x)).sentiment.polarity )
sentiment_summary = data.groupby('cluster')[[f'{col}_sentiment' for col in text_columns]].mean()
print("\nSentiment Analysis Summary:")
sentiment_summary
```
